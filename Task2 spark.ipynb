```python
data_mapped= data.copy()
data_mapped['Species']=data_mapped['Species'].map({'Iris-setosa':0,'Iris-virginica':2,'Iris-versicolor':1})
data_mapped
```

```python
print(data['Species'].unique())
```

```python
x=data_mapped.iloc[:,1:6]
x
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=DeprecationWarning)
warnings.simplefilter(action='ignore', category=RuntimeWarning)
```

```python
kMeans=KMeans(2) 
kMeans.fit(x)
identified_clusters = kMeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters=data_mapped.copy()
data_with_clusters['cluster']=identified_clusters
data_with_clusters
```

```python
kMeans.inertia_
```

```python
wcss=[]

for i in range(1,11):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
    
wcss
```

```python
number_clusters = range(1,11)
plt.plot(number_clusters,wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

```python
kMeans=KMeans(3) 
kMeans.fit(x)
y_clusters = kMeans.fit_predict(x)
y_clusters
```

```python
data_with_clusters=data_mapped.copy()
data_with_clusters['cluster']=y_clusters
data_with_clusters.head(150)
```

```python
plt.scatter(data_with_clusters['SepalLengthCm'],data_with_clusters['SepalWidthCm'],c=data_with_clusters['cluster'],cmap='rainbow')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.show()
```
